# 读取数据

```python
xinwen_path = 'xinwen.txt'
xinwen = list()
with open(xinwen_path,'r',encoding='utf-8') as f:
    for line in f.readlines():
        xinwen.append(line.strip())
print('文本行数:{}'.format(len(xinwen)))
```

# 清洗数据

```python
import re
text = ''
articles = list()
current_numbers = ''
for data in xinwen:
    line = list()
    if data != '':
        articles.append(data)
```

```python
print('文章数量：{}'.format(len(articles)))
```

```python
import random
random.seed(2019) # 设置种子
random.shuffle(articles) # 打乱文本
articles = articles[:30] # 取前30篇文章
```

```python
articles[0]
```

```python
articles_sentences = list()
for article in articles:
    article = re.split('(。|！|\!|\.|？|\?)',article) # 简单分句，保留分割符
    last = ''
    if len(article) % 2 != 0:
        last = article[-1]
    article = [article[2*i]+article[2*i+1] for i in range(int(len(article) / 2))] # 将分隔符与句子拼接
    if last != '':
        article.append(last)
    
    articles_sentences.append(article)
```

```python
print('句子数量：{}'.format(sum([len(article) for article in articles_sentences])))
```

# 分词

```python
import numpy as np
import jiagu as jg
jg.init() # 手动初始化jiagu
```

```python
segment_model_path = 'cnc.model' # 分词模型路径
jg.load_model(segment_model_path) # 使用国家语委分词标准
texts = articles_sentences[0]
articles_words = list()
for article in articles_sentences:
    words = []
    for text in article:
        words.append(jg.cut(text)) # 分词
    articles_words.append(words)
    
for i,article in enumerate(articles_words): 
    print('分词结果 {}:'.format(i))
    for words in article:
        print('{}'.format('/'.join(words)))
    print()
```

# 去除停用词

```python
stopword_path = 'stopwords.txt'
stopword = []

with open(stopword_path,'r',encoding='utf-8') as f:
    for line in f.readlines():
        stopword.append(line[:-1])
print('停用词事例:')
print(random.sample(stopword, 100))

#去除停用词
articles_words_clean = list()

for words in articles_words:
    words_clean = []
    for word in words:
        clean = []
        for w in word:
            if w in stopword:
                continue
            clean.append(w)
        words_clean.append(clean)
    articles_words_clean.append(words_clean)
    
for i,article in enumerate(articles_words_clean):
    print('去停用词结果 {}:'.format(i))
    for words in article:
        print('{}'.format('/'.join(words)))
    print()
```

# 命名实体识别

```python
def get_ner(ner,raw_text):
    entitys = set()
    
    flag_b = False # 标记是否遇到B
    s = -1
    t = ''
    for i,n in enumerate(ner):
        if n == 'O':
            if flag_b:
                entitys.add((raw_text[s:i],t))
                flag_b = False
        else:
            bio,tt = n.split('-')
            if flag_b:
                if bio == 'B':
                    entitys.add((raw_text[s:i],t))
                    flag_b = True
                    s = i
                    t = tt
                else:
                    if tt == t:
                        continue
                    else:
                        entitys.add((raw_text[s:i],t))
                        flag_b = False     
            else:
                if bio == 'B':
                    flag_b = True
                    s = i
                    t = tt
    
    if flag_b:
        entitys.add((raw_text[s:]))
        flag_b = False
        
    return list(entitys)
```

```python
for i,article in enumerate(articles_sentences):
    print('命名实体识别 结果 {}:'.format(i))
    for text in article:
        print('原始文本: {}'.format(text))
        ner = jg.ner(text)
        print('命名实体识别标记: {}'.format(ner))
        ner = get_ner(ner,text)
        print('命名实体识别: {}'.format(ner))
        print()
    print('--------------------------------------------------------------------------------------------------------------------')
    print('--------------------------------------------------------------------------------------------------------------------')
    print()
```

# TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
```

```python
sentence = list()
for articles in articles_words_clean:
    text = list()
    for words in articles:
        text.extend(words)
    text = ' '.join(text)
    sentence.append(text)
print('语料示例: {}'.format(sentence[0]))
```

```python
# 将文本中的词语转换为词频矩阵 
# 该类会将文本中的词语转换为词频矩阵，矩阵元素X[i][j] 表示j词在第i篇文本下的词频  
vectorizer = CountVectorizer()
# 计算每个词语出现的次数  
X = vectorizer.fit_transform(sentence) 
```

```python
# 词袋中所有词语  
word = vectorizer.get_feature_names()  
print(word) 
```

```python
# 查看词频结果
# 矩阵元素X[i][j] 表示j词在第i篇文章下的词频  
print(X.toarray())
```

```python
# 该类会统计每个词语的tf-idf权值  
transformer = TfidfTransformer()  
print(transformer)
print()
# 将词频矩阵X统计成TF-IDF值  
tfidf = transformer.fit_transform(X)  
# 查看数据结构 tfidf[i][j]表示第i篇文章中的tf-idf权重  
print(tfidf.toarray())
```

```python
weight=tfidf.toarray()
# 打印每篇文本的tf-idf词语权重，第一个for遍历所有文章，第二个for遍历某一篇文章下的词语权重  
for i in range(len(weight)):
        print(u"------------这里输出第",i,u"篇文章的词语tf-idf权重------------")  
        for j in range(len(word)):  
            print(word[j],weight[i][j])
        print()
```
